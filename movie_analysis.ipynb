# 🎬 Movie Revenue Analysis

## Objective
Analyze how **budget, genre, and release year** impact movie revenue.

## Dataset
Kaggle Movies Dataset

## Key Questions
- Do higher budgets lead to higher revenue?
- Which genres perform best?

```python
import pandas as pd
```

```python
import ast
```

```python
df = pd.read_csv("movies_metadata.csv", low_memory=False)
```

```python
#Creating a dataframe with the following collumns
```

```python
df = df[["title", "budget", "revenue", "release_date", "genres"]]
```

```python
df.head()
```

```python
# Converting revenue and budget type to number if something cannot be converted it turns it to  NaN(missing value)
```

```python
df["budget"] = pd.to_numeric(df["budget"], errors="coerce")
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")
```

```python
df.head()
```

```python
# keeping only rows where both budget and revenue is greater than 0
```

```python
df = df[(df["budget"]>0) & (df["revenue"]>0)]
```

```python
df.head()
```

```python
#takes the date and converts it to proper date and extracts only the year if value invalid becomes NaT(missing value)
```

```python
df["release_year"] = pd.to_datetime(df["release_date"], errors = "coerce").dt.year
```

```python
#converts text inot a real python object an extracts the first genre from the list of genres in the data and creates a separate column wiht it.
```

```python
def get_genre(x):
    try:
        genres = ast.literal_eval(x)
        return genres[0]["name"] if genres else None
    except:
        return None

df["genre"] = df["genres"].apply(get_genre)
```

```python
df.head()
```

```python
#removing rows where the release year or genre is missing
#then saves data into a csv file
```

```python
df = df.dropna(subset = ["release_year", "genre"])

df.to_csv("clean_movies.csv", index = False)

print("Done! Clean dataset saved as clean_movies.csv")
```
